```python
def get_trial_timecourses_respbased():
    # returns timeourses and means of yes/no/unsure responders separately.
    # doesn't take any inputs, but data_file_loc (location with responseinfo and movie timecourse) 
    # needs to be previously defined
    # Other info: 1. the 'responses' file has a mat of dims nsub*nmovies
    # 2. each movie timecourse file has the dims: nsub*nnodes*ntimepts(32)
    
    responses = np.load(os.path.join(data_file_loc,'responses.npy')) # subs*movies
    nmovies = 10
    for movie_ind in range(nmovies):
        fileName = os.path.join(data_file_loc,'timecourses_run_norm',f'timecourse-all-movies.npy')
        
        movieTC = np.load(fileName) # nsubs * nnodes * ntimepts*nmovies
        movieTC = movieTC[:,:,:,movie_ind]
        
        resp_yes = responses[:,movie_ind] == 1
        resp_no = responses[:,movie_ind] == 0
        resp_unsure = responses[:,movie_ind] == 9
        noresp = np.isnan(responses[:,movie_ind])
        
        if movie_ind == 0:
            movie_tc_yes = np.empty((movieTC.shape[0],movieTC.shape[1],movieTC.shape[2],nmovies))
            movie_tc_no  = np.empty_like(movie_tc_yes)
            movie_tc_unsure  = np.empty_like(movie_tc_yes)
            movie_tc_noresp  = np.empty_like(movie_tc_yes)
            movie_tc_yes[:] = np.nan
            movie_tc_no[:] = np.nan
            movie_tc_unsure[:] = np.nan
            movie_tc_noresp[:] = np.nan
        
        movie_tc_yes[resp_yes,:,:,movie_ind] = movieTC[resp_yes,:,:] # fills out only yes rows
        movie_tc_no[resp_no,:,:,movie_ind] = movieTC[resp_no,:,:] # fills out only no rows
        movie_tc_unsure[resp_unsure,:,:,movie_ind] = movieTC[resp_unsure,:,:] # fills out only unsure rows
        movie_tc_noresp[noresp,:,:,movie_ind] = movieTC[noresp,:,:] # fills out only noresp rows
        # nsubs * nnodes * ntimepts * nmovies
    
    return movie_tc_yes, movie_tc_no, movie_tc_unsure, movie_tc_noresp

# creating the yes/no timecourse files of dim (1049, 268, 32, 10)
movie_tc_yes, movie_tc_no, movie_tc_unsure,movie_tc_noresp = get_trial_timecourses_respbased()
movie_tc_noresp.shape
```

```python
def reshape_tcs_across_movies(movie_tc_yes,movie_tc_no,movie_tc_unsure):
    # reshape the 4dim matrix to a 3dim matrix by fusing subs and movies 
    # Final mat has shape (nsubs*nmovies,nnodes,timepts)
    
    tc_yes_resh = np.empty((int(movie_tc_yes.shape[0]*movie_tc_yes.shape[3]),movie_tc_yes.shape[1],
                                  movie_tc_yes.shape[2]))
    tc_no_resh = np.empty_like(tc_yes_resh)
    tc_unsure_resh = np.empty_like(tc_yes_resh)
    
    sub_id_yes = np.empty((int(movie_tc_yes.shape[0]*movie_tc_yes.shape[3]),)) # this is for bootstrapping 
    # so we can know which subjects were subselected (useful to avoid same subs in multiple groups while 
    # doing ttests between yes and no groups)
    sub_id_no = np.empty_like(sub_id_yes)
    sub_id_unsure = np.empty_like(sub_id_yes)

    for i in range(10):
        mat = movie_tc_yes[:,:,:,i]
        rows = np.arange(i*movie_tc_yes.shape[0],(i+1)*movie_tc_yes.shape[0])
        tc_yes_resh[rows,:,:] = mat
        sub_id_yes[rows] = np.arange(0,len(mat))

    for i in range(10):
        mat = movie_tc_no[:,:,:,i]
        rows = np.arange(i*movie_tc_no.shape[0],(i+1)*movie_tc_no.shape[0])
        tc_no_resh[rows,:,:] = mat 
        sub_id_no[rows] = np.arange(0,len(mat))

    for i in range(10):
        mat = movie_tc_unsure[:,:,:,i]
        rows = np.arange(i*movie_tc_unsure.shape[0],(i+1)*movie_tc_unsure.shape[0])
        tc_unsure_resh[rows,:,:] = mat 
        sub_id_unsure[rows] = np.arange(0,len(mat))


    rows = []
    for i in range(tc_yes_resh.shape[0]): # 0 to 10489
        if len(np.where(~np.isnan(tc_yes_resh[i,:,:].reshape(tc_yes_resh.shape[1]*tc_yes_resh.shape[2],)))[0]) != 0:
            rows.append(i)
    tc_yes_resh = tc_yes_resh[rows,:,:]
    sub_id_yes = sub_id_yes[rows]

    rows = []
    for i in range(tc_no_resh.shape[0]):
        if len(np.where(~np.isnan(tc_no_resh[i,:,:].reshape(tc_no_resh.shape[1]*tc_no_resh.shape[2],)))[0]) != 0:
            rows.append(i)
    tc_no_resh = tc_no_resh[rows,:,:]
    sub_id_no = sub_id_no[rows]

    rows = []
    for i in range(tc_unsure_resh.shape[0]):
        if len(np.where(~np.isnan(tc_unsure_resh[i,:,:].reshape(tc_unsure_resh.shape[1]*tc_unsure_resh.shape[2],)))[0]) != 0:
            rows.append(i)
    tc_unsure_resh = tc_unsure_resh[rows,:,:]
    sub_id_unsure = sub_id_unsure[rows]
    
    return tc_yes_resh,tc_no_resh,tc_unsure_resh

tc_yes_resh,tc_no_resh,tc_unsure_resh = reshape_tcs_across_movies(movie_tc_yes,movie_tc_no,movie_tc_unsure)
tc_yes_resh.shape
```

```python
def fdr_correct(pvals):
    # to fdr-correct pvalues
    
    import mne
    rejected, pval_corr = mne.stats.fdr_correction(pvals, .05)# FDR correction

    return rejected, pval_corr
```

```python
def compare_yn(tyes, tno, fdr=1,test_type='unpaired'):
    # ttests between groups
    # tyes and tno have size 1049*268
    
    t = np.empty((tyes.shape[1],))
    p = np.empty((tyes.shape[1],))
    for n in range(tyes.shape[1]):
        if test_type == 'unpaired':
            t[n],p[n] = stats.ttest_ind(tyes[:,n],tno[:,n], nan_policy='omit',equal_var=False)
        elif test_type == 'paired':
            t[n],p[n] = stats.ttest_rel(tyes[:,n],tno[:,n], nan_policy='omit',equal_var=False)
    if fdr:
        _,p = fdr_correct(p)
    
    return p
```

```python
def end_pts(movie_tc,n_endpts =3):
    #returns the mean of the last "n_endpts" TR from the given timecourse
    return np.nanmean(movie_tc[:,:,-n_endpts:],axis=2) # e.g. n_iter*268
```

```python
# dictionaries to store the last TR, mean differences and corrected p-values of various 
# comparisons between Yes, No and Unsure 


if moviewise:
    print('movie_ind',movie_ind)
    yes_tc = movie_tc_yes[:,:,:,movie_ind]
    no_tc = movie_tc_no[:,:,:,movie_ind]
    unsure_tc = movie_tc_unsure[:,:,:,movie_ind]
else:
    print('across all movies')
    yes_tc = tc_yes_resh
    no_tc = tc_no_resh
    unsure_tc = tc_unsure_resh

timepts = {'yes':end_pts(yes_tc,3),
       'no':end_pts(no_tc,3),
       'unsure':end_pts(unsure_tc,3)} # subj*nodes in each

mean_diffs = {'yes-no':   np.abs(np.nanmean(timepts['yes'],axis=0))-np.abs(np.nanmean(timepts['no'],axis=0)),
             'unsure-no': np.abs(np.nanmean(timepts['unsure'],axis=0))-np.abs(np.nanmean(timepts['no'],axis=0)), \
             'yes-unsure':np.abs(np.nanmean(timepts['yes'],axis=0))-np.abs(np.nanmean(timepts['unsure'],axis=0))}
if moviewise:
    fdr=0
else:
    fdr = 1 # 1:yes to fdr correction of pvalues, 0: uncorrected p-values
    
p_stats = {'yes-no':    compare_yn(timepts['yes'],    timepts['no'],fdr),\
           'unsure-no': compare_yn(timepts['unsure'], timepts['no'],fdr), \
           'yes-unsure':compare_yn(timepts['yes'],    timepts['unsure'],fdr) }


# logicals to test our various hypotheses regarding Yes, No and Unsure to identify various nodes
# the comparisons tested are listed in the dictionary 'comparisons'

comparisons = {0:"Yes>Unsure>No",1:'Yes>[Unsure=No]',2:'No>Unsure>Yes',3: 'No>[Unsure=Yes]',
               4:'[Yes=No]>Unsure',5:'Unsure>[Yes=No]'}#,6:'Unsure=Yes=No'}
logicals ={}
if moviewise and movie_ind==3:
    thresh = .05
else:
    thresh = .05

logicals[comparisons[0]]=[(p_stats['yes-unsure']<=thresh) & (p_stats['unsure-no']<=thresh) & \
                          (p_stats['yes-no']<=thresh) &\
                          (mean_diffs['yes-unsure']>0) & (mean_diffs['unsure-no']>0) &\
                         (mean_diffs['yes-no']>0) ]

logicals[comparisons[1]]=[(p_stats['yes-unsure']<= thresh) & (p_stats['yes-no']<= thresh) & \
                          (p_stats['unsure-no']>thresh) & \
                          (mean_diffs['yes-unsure']>0)& (mean_diffs['yes-no']>0)]

logicals[comparisons[2]]=[(p_stats['unsure-no']<=thresh) & (p_stats['yes-unsure']<=thresh) & \
                          (mean_diffs['unsure-no']<0) & (mean_diffs['yes-unsure']<0)]

logicals[comparisons[3]]=[(p_stats['unsure-no']<=thresh) & (p_stats['yes-no']<=thresh)&
                          (p_stats['yes-unsure']>thresh) & \
                          (mean_diffs['unsure-no']<0) & (mean_diffs['yes-no']<0)]

logicals[comparisons[4]]=[(p_stats['yes-no']> thresh) & (p_stats['yes-unsure']<=thresh) & 
                          (p_stats['unsure-no']<=thresh) & 
                          (mean_diffs['yes-unsure']>0)&(mean_diffs['unsure-no']<0)]

logicals[comparisons[5]]=[(p_stats['yes-unsure']<=thresh)& (p_stats['unsure-no']<=thresh) &\
                          (p_stats['yes-no']> thresh) & \
                          (mean_diffs['yes-unsure']<0)&(mean_diffs['unsure-no']>0)]

#logicals[comparisons[6]]=[(p_stats['yes-unsure']>thresh) & (p_stats['unsure-no']>thresh) &\
#                          (p_stats['yes-no']> thresh)]

```

```python
# dictionaries to store the last TR, mean differences and corrected p-values of various 
# comparisons between Yes, No and Unsure 


comp_type = 1 # enter based on the comparison labels
sig_nodes= comparisons_all.item()['Nodes'][comp_type]
print('nodes:',sig_nodes)
    
first_div = np.empty((n_iter,len(sig_nodes),2)) # stores the earliest location of change
first_div[:]=np.nan
print(first_div.shape)

thresh = .05

for iter in range(n_iter):
    if iter % int(n_iter/10)==0:
        print('Iter',iter)
    if moviewise:
        if iter==0:
            print('movie_ind',movie_ind)
        yes_tc = bootstrapped_iter2(movie_tc_yes,movie_ind,1)
        no_tc = bootstrapped_iter2(movie_tc_no,movie_ind,0)
        unsure_tc = bootstrapped_iter2(movie_tc_unsure,movie_ind,9)
    else:
        if iter==0:
            print('across all movies')
        yes_tc = bootstrapped_iter2(tc_yes_resh)
        no_tc = bootstrapped_iter2(tc_no_resh)
        unsure_tc = bootstrapped_iter2(tc_unsure_resh)

    timepts = {'yes':end_pts(yes_tc,3),'no':end_pts(no_tc,3),
               'unsure':end_pts(unsure_tc,3)} # subj*nodes in each

    mean_diffs = {'yes-no':   np.abs(np.nanmean(timepts['yes'],axis=0))-np.abs(np.nanmean(timepts['no'],axis=0)),
                 'unsure-no': np.abs(np.nanmean(timepts['unsure'],axis=0))-np.abs(np.nanmean(timepts['no'],axis=0)), \
                 'yes-unsure':np.abs(np.nanmean(timepts['yes'],axis=0))-np.abs(np.nanmean(timepts['unsure'],axis=0))}
    
    p_stats = {'yes-no':    compare_yn(timepts['yes'],    timepts['no'],fdr),\
               'unsure-no': compare_yn(timepts['unsure'], timepts['no'],fdr), \
               'yes-unsure':compare_yn(timepts['yes'],    timepts['unsure'],fdr) }


    # logicals to test our various hypotheses regarding Yes, No and Unsure to identify various nodes
    # the comparisons tested are listed in the dictionary 'comparisons'

    logicals ={}
    logicals[comparisons[0]]=[(p_stats['yes-unsure']<=thresh) & (p_stats['unsure-no']<=thresh) & \
                              (p_stats['yes-no']<=thresh) &\
                              (mean_diffs['yes-unsure']>0) & (mean_diffs['unsure-no']>0) &\
                             (mean_diffs['yes-no']>0) ]

    logicals[comparisons[1]]=[(p_stats['yes-unsure']<= thresh) & (p_stats['yes-no']<= thresh) & \
                              (p_stats['unsure-no']>thresh) & \
                              (mean_diffs['yes-unsure']>0)& (mean_diffs['yes-no']>0)]

    logicals[comparisons[2]]=[(p_stats['unsure-no']<=thresh) & (p_stats['yes-unsure']<=thresh) & \
                              (mean_diffs['unsure-no']<0) & (mean_diffs['yes-unsure']<0)]

    logicals[comparisons[3]]=[(p_stats['unsure-no']<=thresh) & (p_stats['yes-no']<=thresh)&
                              (p_stats['yes-unsure']>thresh) & \
                              (mean_diffs['unsure-no']<0) & (mean_diffs['yes-no']<0)]

    logicals[comparisons[4]]=[(p_stats['yes-no']> thresh) & (p_stats['yes-unsure']<=thresh) & 
                              (p_stats['unsure-no']<=thresh) & 
                              (mean_diffs['yes-unsure']>0)&(mean_diffs['unsure-no']<0)]

    logicals[comparisons[5]]=[(p_stats['yes-unsure']<=thresh)& (p_stats['unsure-no']<=thresh) &\
                              (p_stats['yes-no']> thresh) & \
                              (mean_diffs['yes-unsure']<0)&(mean_diffs['unsure-no']>0)]

    
    tscore = np.empty((28,3))
    tscore[:]=np.nan
    pval = np.empty_like(tscore)
    pval[:]=np.nan
    
    if moviewise!=1:
        print('write code for across-movie analysis!!')

    else:#moviewise
        for i,n in enumerate(sig_nodes):
            #print(i)
            for tr in range(28):
                yes_tr = yes_tc[:,n,tr]
                unsure_tr = unsure_tc[:,n,tr]
                no_tr = no_tc[:,n,tr]
                tscore[tr,0],pval[tr,0] = stats.ttest_ind(yes_tr,unsure_tr,nan_policy='omit') # yes v. unsure
                tscore[tr,1],pval[tr,1] = stats.ttest_ind(unsure_tr,no_tr, nan_policy='omit') # unsure v. no
                tscore[tr,2],pval[tr,2] = stats.ttest_ind(yes_tr,no_tr,nan_policy='omit') # yes v. no


            if comp_type == 0:
                good_trs = np.where((pval[:,0]<.05) & (pval[:,1]<.05) & (pval[:,2]<.05) & 
                                    (abs(np.nanmean(yes_tr))>abs(np.nanmean(unsure_tr))) & 
                                    (abs(np.nanmean(unsure_tr))>abs(np.nanmean(no_tr))))[0]
            elif comp_type == 1:
                good_trs = np.where((pval[:,0]<.05) & (pval[:,1]>.05) & (pval[:,2]<.05)  & 
                                    (abs(np.nanmean(yes_tr))>abs(np.nanmean(unsure_tr))) & 
                                    (abs(np.nanmean(yes_tr))>abs(np.nanmean(no_tr))))[0]
            elif comp_type == 2:
                good_trs = np.where((pval[:,0]<.05) & (pval[:,1]<.05) & (pval[:,2]<.05) & 
                                    (abs(np.nanmean(no_tr))>abs(np.nanmean(unsure_tr))) & 
                                    (abs(np.nanmean(unsure_tr))>abs(np.nanmean(yes_tr))))[0]
            elif comp_type == 3:
                good_trs = np.where((pval[:,0]>.05) & (pval[:,1]<.05) & (pval[:,2]<.05) & 
                                    (abs(np.nanmean(no_tr))>abs(np.nanmean(unsure_tr))) & 
                                    (abs(np.nanmean(no_tr))>abs(np.nanmean(yes_tr))))[0]
            elif comp_type == 4:
                good_trs = np.where((pval[:,0]<.05) & (pval[:,1]<.05) & (pval[:,2]>.05) & 
                                    (abs(np.nanmean(yes_tr))>abs(np.nanmean(unsure_tr))) & 
                                    (abs(np.nanmean(no_tr))>abs(np.nanmean(unsure_tr))))[0]
            elif comp_type == 5:
                good_trs = np.where((pval[:,0]<.05) & (pval[:,1]<.05) & (pval[:,2]>.05) &  
                                    (abs(np.nanmean(unsure_tr))>abs(np.nanmean(yes_tr))) & 
                                    (abs(np.nanmean(unsure_tr))>abs(np.nanmean(no_tr))))[0]
            
            #print(good_trs)
            first_div[iter,i,0] = good_trs[0] if len(good_trs)>0 else np.nan
            first_div[iter,i,1] = n
                
                
print('Done')

```
